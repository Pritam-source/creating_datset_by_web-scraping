```python
import pandas as pd
import re
```

```python
path='C:/Github_clones/creating_datset_by_web-scraping/Disney_movies.csv'
```

```python
data = pd.read_csv(path) 
data.head()
```

```python
data.drop("Unnamed: 0",inplace=True, axis=1)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.fillna("N/A", inplace = True)
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
print(type(data["Running_time"]))
```

```python
data["Running_time(in minutes)"]= data["Running_time"].str.replace('min',"")
data["Running_time(in minutes)"]= data["Running_time(in minutes)"].str.replace('utes',"")
```

```python
data.head()
```

```python
data["Box_office"]= data["Box_office"].str.replace('million',"M")
data["Box_office"]= data["Box_office"].str.replace('',"M")
```

```python
data.head()
```

```python
clutter_one = re.compile(r'\(.*\)$')
data["Box_office"]= data["Box_office"].str.replace(clutter_one,"")
```

```python
data.head()
```

```python
data["Budget"]= data["Budget"].str.replace("million","M")
```

```python
data.head()
```

```python
data.rename(columns = {'Budget':'Budget(USD)','Box_office':'Box_office(USD)'}, inplace = True)
```

```python
data.head()
```

```python
data["Budget(USD)"]= data["Budget(USD)"].str.replace("$","")
data["Box_office(USD)"]= data["Box_office(USD)"].str.replace("$","")
```

```python
data.head()
```

```python
data["Box_office(USD)"]= data["Box_office(USD)"].str.replace(">","")
```

```python
data["Running_time(in minutes)"]= data["Running_time(in minutes)"].str.replace(clutter_one,"")
```

```python
data["Box_office(USD)"]= data["Box_office(USD)"].str.replace(clutter_one,"")
data["Budget(USD)"]= data["Budget(USD)"].str.replace(clutter_one,"")
```

```python
data.loc[187,:]
```

```python
data.drop("Running_time",inplace=True, axis=1)
```

```python
data.head()
```

```python
data["Box_office(USD)"]= data["Box_office(USD)"].str.replace("billion","B")
```

```python
data.to_excel("cleaned_Disney_movies_r.xlsx")
```
